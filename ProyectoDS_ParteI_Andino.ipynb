###Rating basado en el tipo de cocina:

```python
#Crear la figura
sns.boxplot(x=df["Cuisine"], y=df["Rating"])

#Añadir el titulo
plt.title('Distribución de Rating por Tipo de Cocina')

#Mostrar el gráfico
plt.show()
```

Existe un claro equilibrio del raiting entre los distintos tipos de cocina ofrecidos. Esto se puede traducir en que los clientes no tienen preferencia por algún tipo de cocina en específico.

###Rating y disponibilidad de estacionamiento:


```python
# Crear la figura y los subplots
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar la densidad para "Yes"
sns.kdeplot(df[df['Parking Availability'] == 'Yes']['Rating'], label='Yes', color='blue', ax=ax, fill=True)

# Graficar la densidad para "No"
sns.kdeplot(df[df['Parking Availability'] == 'No']['Rating'], label='No', color='orange', ax=ax, fill=True)

# Añadir etiquetas y título
ax.set_xlabel('Rating')
ax.set_ylabel('Densidad')
ax.set_title('Gráfico de Densidad de Rating según Disponibilidad de Estacionamiento')

# Añadir la leyenda
ax.legend(title='Parking Availability')

# Agregar grilla y Mostrar el gráfico
plt.grid(True)
plt.show()
```

Existe un claro equilibrio del raiting según la disponibilidad de estacionamiento propio para los clientes. Esto se puede traducir en que los clientes no consideran esta característica a la hora de evaluar el restaurante.

###Rating y presupuesto de marketing:



```python
# Crear la figura y el gráfico lineal
plt.figure(figsize=(8, 6))

# Graficar la relación entre Rating y Marketing Budget
plt.plot(df['Marketing Budget'], df['Rating'], marker='o', linestyle='-', color='green')

# Añadir etiquetas y título
plt.xlabel('Marketing Budget')
plt.ylabel('Rating')
plt.title('Relación entre Rating y Marketing Budget')

# Agregar grilla y Mostrar el gráfico
plt.grid(True)
plt.show()
```

Es claro que, a partir de una inversion de 6000 es mayor el raiting obtenido en una correlación lineal. Sin embargo, en montos menores de inversión también es posible obtener puntuaciones altas.
En conclusión, existe una correlación entre el rating y la inversion en marketing pero no es muy clara.

###Rating y la calidad del servicio:

```python
from scipy.stats import pearsonr
# Calcular el coeficiente de correlación de Pearson
corr, _ = pearsonr(df['Rating'], df['Service Quality Score'])

# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Rating', y='Service Quality Score', data=df, color='purple', s=70)

# Añadir el coeficiente de correlación en el título
plt.title(f'Correlograma entre Rating y Service Quality Score (r = {corr:.2f})')

# Añadir etiquetas de los ejes
plt.xlabel('Rating')
plt.ylabel('Service Quality Score')

# Agregar grilla y Mostrar el gráfico
plt.grid(True)
plt.show()
```

Existe una distribución llamativamente homogénea entre los valores de rating en base a la calidad del servicio, por lo que podemos concluir que no existe correlación entre ellas.

###Rating y la experiencia del Chef:

```python
# Generar las variables
x= df['Chef Experience Years']
y= df['Rating']

# Inicializar el layout
fig, ax = plt.subplots(figsize=(9, 9))

# Agregar el scatterplot
ax.scatter(x, y, s=60, alpha=0.7, edgecolors="red")

# Añadir etiquetas y título
ax.set_xlabel('Chef Experience Years')
ax.set_ylabel('Rating')
ax.set_title('Relación entre experiencia del Chef y Rating')

# Añadir la línea de regresión
m, b = np.polyfit(x, y, 1)
ax.plot(x, m*x + b, color="k", lw=2.5)

# Agregar grilla y Mostrar el gráfico
plt.grid(True)
plt.show()
```

Existe una distribución llamativamente homogénea entre los valores de rating en base a los años de experiencia del Chef, por lo que podemos concluir que no existe correlación entre ellas.

###Correlación entre factores:

```python
# Seleccionar solo las columnas numéricas del DataFrame
df_numeric = df.select_dtypes(include='number')

# Calcular la matriz de correlación
corr_matrix = df_numeric.corr()

# Crear el heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Añadir el título
plt.title('Heatmap de Correlación entre Variables Numéricas')

# Mostrar el gráfico
plt.show()
```

Debido a que los análisis anteriores no han concludio en un factor que afecte al rating, se realizó un análisis general incluyendo todas las variables para encontrar una respuesta mejor orientada.
En la gráfica, podemos visualizar que el rating se ve influenciado en su mayoria por el presupuesto de marketing y la cantidad de seguidores en redes sociales y, en menor medida, por los años de experiencia del chef y el puntaje de ambiente
